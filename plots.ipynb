```python
from numpy import arange, array, cos, linspace, pi, sign, sin, sum, append, zeros
import matplotlib.pyplot as plt


def main():
    Xs = linspace(-10, 10, num=1000)
    y_true = square(Xs, 1)

    y_pred = array([])

    an = zeros(3)
    bn = squareParamtest(3, 1)
    bn[1::2] = 0

    for x in Xs:
        y = fourier(x, 0, 0.5, an, bn)
        y_pred = append(y_pred, y)

    plt.plot(Xs, y_pred)
    plt.plot(Xs, y_true)


def square(x, period):
    return sign(sin((2 * pi * x) / period))


def fourier(x, a0, L, an, bn):
    na = arange(1, len(an) + 1)
    nb = arange(1, len(bn) + 1)
    return a0 + sum(an * cos(na * x * (pi / L))) + sum(bn * sin(nb * x * (pi / L)))


def squareParamtest(n, h):
    p = array([])
    for i in range(1, n + 1):
        v = (4 * h) / (i * pi)
        p = append(p, v)
    return p


if __name__ == "__main__":
    main()

```
